```python
import cv2

image_path = "test/images/Day-0-Col-R1H1C1_jpg.rf.b7ba961c04955b91ed72300d49e91637.jpg"
# Obtener la imagen original
image = cv2.imread(image_path)  # Leer la imagen con OpenCV para poder modificarla

# Obtener las cajas delimitadoras y calcular el área de cada una
detections = results[0].boxes  # Obtiene las cajas detectadas
areas = []

# Parámetros para el texto
font_scale = 1.2  # Ajustar este valor para aumentar el tamaño de la letra
color = (255, 0, 0)  # Color del texto en BGR
thickness = 2  # Grosor del texto
box_color = (0, 255, 0)  # Color de la caja en BGR (verde)

# Dibujar las cajas y las áreas en la imagen
for box in detections:
    # Coordenadas de cada caja
    x_min, y_min, x_max, y_max = map(int, box.xyxy[0])  # Convertir a enteros para opencv
    width = x_max - x_min
    height = y_max - y_min
    area = width * height
    areas.append(area)
    
    # Dibujar la caja sobre la imagen
    cv2.rectangle(image, (x_min, y_min), (x_max, y_max), box_color, 2)
    
    # Definir el texto que muestra el área
    area_text = f"Area: {area:.2f} px^2"
    
    # Colocar el texto sobre la imagen
    cv2.putText(image, area_text, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 
                font_scale, color, thickness)

# Mostrar la imagen con Matplotlib
plt.figure(figsize=(10, 8))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convertir BGR a RGB para mostrar correctamente
plt.axis("off")
plt.show()
```
